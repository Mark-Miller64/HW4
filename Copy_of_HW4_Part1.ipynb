```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

**Q1** Please use Pandas to read olympic_medals.csv and use parallel_categories function from plotly.express to visualize proportions of medal type for each gender from since year 2000. Please see the example in the Python notebook we walked through in the class.

```python
#please use this cell to read and select your data
df = pd.read_csv('/content/drive/MyDrive/DATA/olympic_medals.csv')
df
df = df[df['Year'] >= 2000]
df
df_counts = df.groupby(['Year', 'Gender', 'Medal']).size().reset_index(name='Count')
```

```python
#Please use this cell to create your your figure. Please use Year column to color your graph.

fig = px.parallel_categories(
    df_counts,
    dimensions=['Year', 'Gender', 'Medal'],
    color='Year',
    color_continuous_scale=px.colors.sequential.Viridis,
    labels={'Year': 'Year', 'Gender': 'Gender', 'Medal': 'Medal', 'Count': 'Count'},
    title="Proportions of Medal Types by Gender Since 2000"
)

# Show the plot
fig.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# create DataFrame
df = pd.DataFrame({'High Temp': [28, 30, 34, 38, 45, 42,
                                 38, 35, 32, 28, 25, 21],
                   'Low Temp': [22, 26, 30, 32, 41, 38,
                                32, 31, 28, 22, 15, 15],
                   'Avg Temp': [25, 28, 32, 35, 43, 40,
                                35, 33, 30, 25, 20, 18]},
                  index=['Jan', 'Feb', 'Mar', 'Apr', 'May',
                         'Jun', 'Jul', 'Aug', 'Sep', 'Oct',
                         'Nov', 'Dec'])


# create stacked bar chart for monthly temperatures
df.plot(kind='bar', stacked=True, color=['red', 'skyblue', 'green'])

# labels for x & y axis
plt.xlabel('Months')
plt.ylabel('Temp ranges in Degree Celcius')

# title of plot
plt.title('Monthly Temperatures in a year')
```

```python
#You can use this cell to write your code. It is doable at most 4 lines of code.
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/DATA/2016elections.csv')

# List State ABV. to filter by
states_abv = ['AR', 'MI', 'CA', 'WI']

# Filter based on the 'st' column
filtered_df = df[df['st'].isin(states_abv)]

# Prepare the data for plotting
pivot_df = filtered_df.pivot_table(index='st',
                                    values=['pct_clinton','pct_trump', 'pct_johnson','pct_other'],
                                    aggfunc='mean')  # Use mean if there are multiple rows per state

pivot_df = pivot_df.rename(columns={'pct_clinton': 'Clinton',
                                     'pct_trump': 'Trump',
                                     'pct_johnson': 'Johnson',
                                     'pct_other': 'Other'})


# Plot the stacked bar chart
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 7), colormap='plasma')

# Set labels and title
plt.xlabel('State')
plt.ylabel('Percentage')
plt.title('Election Percentages by State')
plt.legend(title='Candidates', loc='upper right')
plt.tight_layout()

# Show the plot
plt.show()




```
